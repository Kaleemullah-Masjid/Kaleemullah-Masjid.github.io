```python
import sqlite3

# SQLite database connection
conn = sqlite3.connect('Weather_Data.db')  # Replace 'your_database.db' with the name of your SQLite database

# Create a cursor object to execute SQL statements
cursor = conn.cursor()

# Specify the SQL statement to create a table
create_table_sql = '''
CREATE TABLE IF NOT EXISTS Weather_Forcast (
    name TEXT
    ,startTime TEXT
    ,endTime TEXT
    ,isDaytime TEXT
    ,temperature REAL
    ,temperatureUnit TEXT
    ,temperatureTrend TEXT
    ,windSpeed TEXT
    ,windDirection TEXT
    ,shortForecast TEXT
    ,detailedForecast TEXT
    ,probabilityOfPrecipitation TEXT
    ,dewpoint REAL
    ,relativeHumidity REAL
    ,date TEXT
    ,Year INT
    ,Month INT
    ,Day INT
    ,Hour INT
    ,Date_Created TEXT

);
'''

# Execute the SQL statement to create the table
cursor.execute(create_table_sql)

# Commit the changes and close the connection
conn.commit()
conn.close()
```

```python
engine = create_engine('sqlite:///Weather_Data.db')
df = pd.read_sql_query(f"SELECT * From Weather_Forcast", engine)
```

```python
import numpy as np
df['Hour_Start']=df.startTime.str.split(' ',expand=True)[1].str.split('-',expand=True)[0].str.split('.',expand=True)[0].str.split(':',expand=True)[0]
df['Hour_End']=df.endTime.str.split(' ',expand=True)[1].str.split('-',expand=True)[0].str.split('.',expand=True)[0].str.split(':',expand=True)[0]
df['Night'] = np.where(df['Hour_Start'].astype('int') < 18, 0, 1)
df['Date_Time'] = pd.concat([
df.Year.astype('str').str.rjust(4,'0') 
, df.Month.astype('str').str.rjust(2,'0')
, df.Day.astype('str').str.rjust(2,'0')
, df.Night.astype('str').str.rjust(2,'0')
], axis = 1).apply(lambda x:'_'.join(x),axis=1)

df.sort_values('Date_Time')
```
